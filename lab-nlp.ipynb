```python
#function to cleaning, tokenizing, stemming, lemmatizing and removing stopwords

import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer, WordNetLemmatizer

# Precompile regular expressions to gain efficiency
url_pattern = re.compile(r'https?://\S+|www\.\S+')
symbols_pattern = re.compile(r'[^a-zA-Z\s]')

# Define stopwords as a set
stop_words = set(stopwords.words('english'))

def tokenize_stem_lemmatize_stopwords(text):
    # Remove URLs
    text_without_urls = url_pattern.sub('', text)
    
    # Remove symbols
    cleaned_text = symbols_pattern.sub(' ', text_without_urls)
    
    # Tokenize
    tokens = word_tokenize(cleaned_text.lower())  # Tokenize and convert to lowercase
    
    # Stemming
    ps = SnowballStemmer(language='english')
    stemmed = [ps.stem(word) for word in tokens]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(word) for word in stemmed]
    
    # Remove stopwords
    without_sw = [word for word in lemmatized if word not in stop_words]
    
    return without_sw
```

```python
#importing dataset

tweets = pd.read_csv('tweets.csv', encoding='ISO-8859-1')
tweets.head()
```

```python
tweets.columns = ['target','id','date','flag','user','text']
```

```python
tweets
```

```python
#creating a sample of 20000

sampled_tweets = tweets.sample(n=20000)
sampled_tweets = sampled_tweets.reset_index()
sampled_tweets
```

```python
#applying previous function to the sample

sampled_tweets['text_processed'] = sampled_tweets['text'].apply(tokenize_stem_lemmatize_stopwords)
sampled_tweets
```

```python
def re_blob(row):
    return " ".join(row['text_processed'])

sampled_tweets['text_processed'] = sampled_tweets.apply(re_blob,axis=1)
sampled_tweets.head()
```

```python
neg_tweets = sampled_tweets[sampled_tweets['target']==0]
neg_tweets
```

```python
pos_tweets = sampled_tweets[sampled_tweets['target']==4]
pos_tweets
```

```python
#let's take only the most common 1000 words
bow_vect = CountVectorizer(max_features=1000)
# fit creates one entry for each different word seen
X = bow_vect.fit_transform(sampled_tweets['text_processed']).toarray()
```

```python
as_df = pd.DataFrame(X,columns=bow_vect.get_feature_names_out())
as_df.head(10)
```

```python
#dividing tweets into two clusters

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
pred = kmeans.predict(X)
```

```python
predict_df = pd.concat([sampled_tweets['text'],pd.DataFrame(pred,columns=['class'])],axis=1)
predict_df.head()
```
